```python
from dota2_api import get_api_json
import pandas as pd
import pickle
import datetime


with open('hero_name.pickle', 'rb') as handle:  
    hero_name = pickle.load(handle)

sql = """SELECT 
league_match_info.*,
teams.name as team_name,
teams.tag as team_tag
from 
(SELECT
leagues.name as league_name,
leagues.leagueid,
matches.match_id,
matches.start_time,
EXTRACT(epoch FROM age(current_timestamp, to_timestamp(matches.start_time)))/3600 as hours_to_now,
matches.radiant_win,
matches.duration,
matches.cluster,
matches.lobby_type,
matches.game_mode,
matches.engine,
notable_players.name as player_name,
player_matches.hero_id,
player_matches.kills,
player_matches.deaths,
player_matches.assists,
player_matches.last_hits,
player_matches.gold_per_min,
player_matches.xp_per_min,
CASE WHEN player_matches.player_slot < 5 THEN matches.radiant_team_id ELSE matches.dire_team_id END as team_id,
CASE WHEN player_matches.player_slot < 5 THEN 'radiant' ELSE 'dire' END as camp,
CASE WHEN (player_matches.player_slot < 5 and matches.radiant_win = true) or (player_matches.player_slot >= 5 and matches.radiant_win = false) THEN true ELSE false END as is_win
FROM matches
JOIN leagues using(leagueid)
JOIN player_matches using(match_id)
JOIN notable_players using(account_id)
WHERE leagueid = 9943  and start_time > 1527904857
) league_match_info
JOIN teams using(team_id)
order by start_time

"""

# WHERE leagueid = 9943  and start_time > 1527904857
# leagueid用来筛选联赛，start_time用来筛选正赛（排除预选赛）
# 可以到dotabuff.com查看id和正赛开始的时间


matches = pd.DataFrame(get_api_json('https://api.opendota.com/api/explorer?sql={}'.format(sql))['rows'])
cn_matches = matches.copy()
cn_matches['hero_id'] = cn_matches['hero_id'].apply(lambda x:hero_name[x][1])
# cn_matches
matches
```

```python
#单场比赛最高\低
for index, ascending in [['kills', False], ['assists', False], ['deaths', False], ['last_hits', False], ['gold_per_min', False]]:
    print(matches.sort_values(index, ascending=ascending).iloc[0:3][['team_name', 'player_name', index]])
```

```python
#选手平均比赛最高\低
for index, ascending in [['kills', False], ['assists', False], ['deaths', True], ['last_hits', False], ['gold_per_min', False]]:
    print(matches.groupby('player_name')[index].mean().sort_values(ascending=ascending)[0:3])
```

```python
#英雄个数
print(matches.groupby('player_name')['hero_id'].nunique().sort_values(ascending=False)[0:3])
#单场比赛最高双方总人头
print(matches.groupby('match_id')['kills'].sum().sort_values(ascending=False)[0:3])
#最长\短比赛时间
print(pd.DataFrame(matches.groupby('match_id')['duration'].mean().sort_values(ascending=False)[0:5]).join(
    matches[['match_id', 'is_win', 'team_name']].set_index('match_id'),  lsuffix='_caller', rsuffix='_other').reset_index().groupby(['match_id'])[['duration', 'team_name', 'is_win']].max().sort_values('duration', ascending=False))
print(pd.DataFrame(matches.groupby('match_id')['duration'].mean().sort_values(ascending=True)[0:5]).join(
    matches[['match_id', 'is_win', 'team_name']].set_index('match_id'),  lsuffix='_caller', rsuffix='_other').reset_index().groupby(['match_id'])[['duration', 'team_name', 'is_win']].max().sort_values('duration', ascending=True))
```

```python
#单场比赛总XX最高的队伍
for index, ascending in [['kills', False], ['deaths', True], ['assists', False]]:
    print(matches.groupby(['team_name', 'match_id'])[index].sum().sort_values(ascending=ascending)[0:3])
    
#场均比赛XX最高的队伍
print(pd.DataFrame(matches.groupby(['team_name', 'match_id'])['kills'].sum()).reset_index().groupby('team_name')['kills'].mean().sort_values(ascending=ascending)[0:3])
print(pd.DataFrame(matches.groupby(['team_name', 'match_id'])['duration'].max()).reset_index().groupby('team_name')['duration'].mean().sort_values(ascending=ascending)[0:3])

#使用英雄数量最多\少的队伍
print(matches.groupby('team_name')['hero_id'].nunique().sort_values(ascending=False)[0:3])
pd.DataFrame(matches.groupby('team_name')['hero_id'].nunique().sort_values(ascending=True)[0:8]).join(pd.DataFrame(matches.groupby('team_name')['match_id'].nunique()))
```

```python
#上场\禁用次数最多的英雄
print(cn_matches.groupby('hero_id')['match_id'].nunique().sort_values(ascending=False)[0:3])

#英雄平均比赛最高\低
for index, ascending in [['kills', False], ['assists', False], ['deaths', True], ['last_hits', False], ['xp_per_min', False]]:
    print(pd.DataFrame(cn_matches.groupby('hero_id')[index].mean().sort_values(ascending=ascending)[0:8]).join(
    pd.DataFrame(cn_matches.groupby('hero_id')['match_id'].nunique().sort_values(ascending=ascending))))
    
#英雄单场最高
print(cn_matches.sort_values('kills', ascending=False).iloc[0:5][['hero_id', 'player_name', 'kills']])
print(cn_matches.sort_values('last_hits', ascending=False).iloc[0:5][['hero_id', 'player_name', 'last_hits']])
```
